```python
import os
import numpy as np
import torchani
from torchani.data._pyanitools import anidataloader
from torchani.data._pyanitools import datapacker

mol_data = dict()
n_mol = 0
db = anidataloader('./ani1x-20191001_rz.h5')
dpack1 = datapacker('./ani1x-20191001_wb97x_dz_dipole.h5')
dpack2 = datapacker('./ani1x-20191001_wb97x_tz_dipole.h5')
dpack3 = datapacker('./ani1x-20191001_ccsdt_cbs.h5')

for i, data in enumerate(db):
    # print(data['path'])
    # print(data.keys())
    n_mol += 1
    species = data['atomic_numbers']
    coordinates = data['coordinates']
    wb97x_dz_dipole = data['wb97x_dz.dipole']
    wb97x_dz_cm5 = data['wb97x_dz.cm5_charges']
    wb97x_dz_hirshfeld = data['wb97x_dz.hirshfeld_charges']
    wb97x_dz_energy = data['wb97x_dz.energy']
    wb97x_dz_forces = data['wb97x_dz.forces']
    wb97x_tz_dipole = data['wb97x_tz.dipole']
    wb97x_tz_energy = data['wb97x_tz.energy']
    wb97x_tz_forces = data['wb97x_tz.forces']
    wb97x_tz_charges = data['wb97x_tz.mbis_charges']
    ccsdt_cbs_energy = data['ccsd(t)_cbs.energy']
    
    wb97x_dz_index = np.where(np.isfinite(wb97x_dz_dipole))
    wb97x_dz_index1 = np.unique(wb97x_dz_index[0])
    wb97x_dz_coordinates = coordinates[wb97x_dz_index1, :, :]
    wb97x_dz_dipole = wb97x_dz_dipole[wb97x_dz_index1, :]
    wb97x_dz_cm5 = wb97x_dz_cm5[wb97x_dz_index1, :]
    wb97x_dz_hirshfeld = wb97x_dz_hirshfeld[wb97x_dz_index1, :]
    wb97x_dz_energy = wb97x_dz_energy[wb97x_dz_index1]
    wb97x_dz_forces = wb97x_dz_forces[wb97x_dz_index1, :, :]
    
    wb97x_tz_index = np.where(np.isfinite(wb97x_tz_dipole))
    wb97x_tz_index1 = np.unique(wb97x_tz_index[0])
    wb97x_tz_coordinates = coordinates[wb97x_tz_index1, :, :]
    wb97x_tz_dipole = wb97x_tz_dipole[wb97x_tz_index1, :]
    wb97x_tz_energy = wb97x_tz_energy[wb97x_tz_index1]
    wb97x_tz_forces = wb97x_tz_forces[wb97x_tz_index1, :, :]
    wb97x_tz_charges = wb97x_tz_charges[wb97x_tz_index1, :]
    
    ccsdt_cbs_index = np.where(np.isfinite(ccsdt_cbs_energy))
    ccsdt_cbs_coordinates = coordinates[ccsdt_cbs_index, :, :]
    ccsdt_cbs_energy = ccsdt_cbs_energy[ccsdt_cbs_index]

    dpack1.store_data(data['path'],
                      species = species,
                      coordinates = wb97x_dz_coordinates,
                      energies = wb97x_dz_energy,
                      dipoles = wb97x_dz_dipole,
                      forces = wb97x_dz_forces,
                      cm5 = wb97x_dz_cm5,
                      hirshfeld = wb97x_dz_hirshfeld,
                     )  

    dpack2.store_data(data['path'],
                      species = species,
                      coordinates = wb97x_tz_coordinates,
                      energies = wb97x_tz_energy,
                      dipoles = wb97x_tz_dipole,
                      forces = wb97x_tz_forces,
                      charges = wb97x_tz_charges,
                     )
    
    dpack3.store_data(data['path'],
                      species = species,
                      coordinates = ccsdt_cbs_coordinates,
                      energies = ccsdt_cbs_energy,
                     )
    
dpack1.cleanup()
dpack2.cleanup()
dpack3.cleanup()
```
