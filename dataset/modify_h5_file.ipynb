```python
import os
import numpy as np
import torchani
from torchani.data._pyanitools import anidataloader
from torchani.data._pyanitools import datapacker
```

```python
db = anidataloader('drugbank_testset.h5')
mol_data = dict()
n_mol = 0
for i, data in enumerate(db):
    print(data['path'])
    print(data.keys())
    n_mol += 1
    mol_data[str(n_mol)] = data
    if i == 1:
        print(data['species'])
        # print(data['energies'])
        # print(data['cm5'])
        mol = data['species']
        unique, counts = np.unique(mol, return_counts=True)
        print(unique, counts)
```

```python
# print(mol_data['1'])
print(len(mol_data['1']['species']))
print(mol_data['1']['coordinates'].shape)
print(mol_data['1']['energies'].shape)
print(mol_data['1']['cm5'].shape)
print(mol_data['1']['forces'].shape)
print(mol_data['1']['hirdipole'].shape)
print(mol_data['1']['hirshfeld'].shape)
print(mol_data['1']['spindensities'].shape)
sidx = np.arange(55)
print(sidx)
mol_data['1']['cm5'][:, 55]
```

```python
dpack = datapacker('drugbank_testset_mod.h5')
```

```python
for k, data in mol_data.items():
    Na = len(data['species'])
    data_mod = {'species': data['species'],
                'coordinates': data['coordinates'],
                'energies': data['energies'],
                'forces': data['forces'],
                'cm5': data['cm5'][:, 0:Na],
                'cm5_total': data['cm5'][:, Na],
                'hirshfeld': data['hirshfeld'][:, 0:Na],
                'hirshfeld_total': data['hirshfeld'][:, Na]
                }

    dpack.store_data(data['path'],
                     species = data_mod['species'],
                     coordinates = data_mod['coordinates'],
                     energies = data_mod['energies'],
                     forces = data_mod['forces'],
                     cm5 = data_mod['cm5'],
                     cm5_total = data_mod['cm5_total'],
                     hirshfeld = data_mod['hirshfeld'],
                     hirshfeld_total = data_mod['hirshfeld_total'],
                    )        
dpack.cleanup()
```
