```python
import os
import numpy as np
import torchani
from torchani.data._pyanitools import anidataloader
from torchani.data._pyanitools import datapacker
import collections
```

```python
db = anidataloader('./COMP6/COMP6v1/DrugBank/drugbank_testset.h5')
mol_data = collections.OrderedDict()
n_mol = 0
for i, data in enumerate(db):
    # print(data['path'])
    # print(data.keys())
    n_mol += 1
    mol_data[str(n_mol)] = data
    if i == 1:
        print(data['path'])
        print(data.keys())
        print(data['species'])
        # print(data['energies'])
        # print(data['cm5'])
        mol = data['species']
        unique, counts = np.unique(mol, return_counts=True)
        print(unique, counts)
```

```python
len_train = len(mol_data) - len(mol_data)//5  # 80% for training
train_mol_data = collections.OrderedDict(list(mol_data.items())[:len_train])
valid_mol_data = collections.OrderedDict(list(mol_data.items())[len_train:])
```

```python
train_dpack = datapacker('./COMP6/COMP6v1/DrugBank/drugbank_testset_mod_training.h5')
valid_dpack = datapacker('./COMP6/COMP6v1/DrugBank/drugbank_testset_mod_validation.h5')

for k, data in train_mol_data.items():
    Na = len(data['species'])
    data_mod = {'species': data['species'],
                'coordinates': data['coordinates'],
                'energies': data['energies'],
                'forces': data['forces'],
                'cm5': data['cm5'][:, 0:Na],
                'cm5_total': data['cm5'][:, Na],
                'hirshfeld': data['hirshfeld'][:, 0:Na],
                'hirshfeld_total': data['hirshfeld'][:, Na]
                }

    train_dpack.store_data(data['path'],
                     species = data_mod['species'],
                     coordinates = data_mod['coordinates'],
                     energies = data_mod['energies'],
                     forces = data_mod['forces'],
                     cm5 = data_mod['cm5'],
                     cm5_total = data_mod['cm5_total'],
                     hirshfeld = data_mod['hirshfeld'],
                     hirshfeld_total = data_mod['hirshfeld_total'],
                    )        
train_dpack.cleanup()

for k, data in valid_mol_data.items():
    Na = len(data['species'])
    data_mod = {'species': data['species'],
                'coordinates': data['coordinates'],
                'energies': data['energies'],
                'forces': data['forces'],
                'cm5': data['cm5'][:, 0:Na],
                'cm5_total': data['cm5'][:, Na],
                'hirshfeld': data['hirshfeld'][:, 0:Na],
                'hirshfeld_total': data['hirshfeld'][:, Na]
                }

    valid_dpack.store_data(data['path'],
                     species = data_mod['species'],
                     coordinates = data_mod['coordinates'],
                     energies = data_mod['energies'],
                     forces = data_mod['forces'],
                     cm5 = data_mod['cm5'],
                     cm5_total = data_mod['cm5_total'],
                     hirshfeld = data_mod['hirshfeld'],
                     hirshfeld_total = data_mod['hirshfeld_total'],
                    )        
valid_dpack.cleanup()
```

```python
train_dpack.cleanup()
valid_dpack.cleanup()
```

```python
db = anidataloader('./COMP6/COMP6v1/DrugBank/drugbank_testset.h5')
mol_data = collections.OrderedDict()
n_mol = 0
for i, data in enumerate(db):
    # print(data['path'])
    # print(data.keys())
    n_mol += 1
    mol_data[str(n_mol)] = data
    if i == 1:
        print(data['path'])
        print(data.keys())
        print(data['species'])
        # print(data['energies'])
        # print(data['cm5'])
        mol = data['species']
        unique, counts = np.unique(mol, return_counts=True)
        print(unique, counts)


len_train = len(mol_data) - len(mol_data)//5  # 80% for training
train_mol_data = collections.OrderedDict(list(mol_data.items())[:len_train])
valid_mol_data = collections.OrderedDict(list(mol_data.items())[len_train:])

train_dpack = datapacker('./COMP6/COMP6v1/DrugBank/drugbank_testset_mod2_training.h5')
valid_dpack = datapacker('./COMP6/COMP6v1/DrugBank/drugbank_testset_mod2_validation.h5')

for k, data in train_mol_data.items():
    Na = len(data['species'])
    data_mod = {'species': data['species'],
                'coordinates': data['coordinates'],
                'energies': data['energies'],
                'forces': data['forces'],
                'cm5': data['cm5'][:, 0:Na],
                'cm5_total': data['cm5'][:, Na],
                'hirshfeld': data['hirshfeld'][:, 0:Na],
                'hirshfeld_total': data['hirshfeld'][:, Na],
                'hirdipole': data['hirdipole'][:, 0:Na, :],
                'dipole': data['hirdipole'][:, Na, :]
                }

    train_dpack.store_data(data['path'],
                     species = data_mod['species'],
                     coordinates = data_mod['coordinates'],
                     energies = data_mod['energies'],
                     forces = data_mod['forces'],
                     cm5 = data_mod['cm5'],
                     cm5_total = data_mod['cm5_total'],
                     hirshfeld = data_mod['hirshfeld'],
                     hirshfeld_total = data_mod['hirshfeld_total'],
                     hirdipole = data_mod['hirdipole'],
                     dipole = data_mod['dipole']
                    )        
train_dpack.cleanup()

for k, data in valid_mol_data.items():
    Na = len(data['species'])
    data_mod = {'species': data['species'],
                'coordinates': data['coordinates'],
                'energies': data['energies'],
                'forces': data['forces'],
                'cm5': data['cm5'][:, 0:Na],
                'cm5_total': data['cm5'][:, Na],
                'hirshfeld': data['hirshfeld'][:, 0:Na],
                'hirshfeld_total': data['hirshfeld'][:, Na],
                'hirdipole': data['hirdipole'][:, 0:Na, :],
                'dipole': data['hirdipole'][:, Na, :]
                }

    valid_dpack.store_data(data['path'],
                     species = data_mod['species'],
                     coordinates = data_mod['coordinates'],
                     energies = data_mod['energies'],
                     forces = data_mod['forces'],
                     cm5 = data_mod['cm5'],
                     cm5_total = data_mod['cm5_total'],
                     hirshfeld = data_mod['hirshfeld'],
                     hirshfeld_total = data_mod['hirshfeld_total'],
                     hirdipole = data_mod['hirdipole'],
                     dipole = data_mod['dipole']
                    )        
valid_dpack.cleanup()
```
