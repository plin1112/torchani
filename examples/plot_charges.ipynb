```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

lines = np.loadtxt("comparison_cm5_calculated_vs_predicted.txt", dtype=np.float64, usecols=(1,2), unpack=False)
x = lines[:, 0]
y = lines[:, 1]

plt.figure(0)
plt.clf()
# plt.scatter(x, y, edgecolors='none', s=4, alpha=0.5)
# plt.show()

A = np.vstack([x, np.ones(len(x))]).T
model, resid = np.linalg.lstsq(A, y, rcond=None)[:2]
m, c = model
r2 = 1 - resid / (y.size * y.var())
rmse = np.linalg.norm(m*x + c - y) / np.sqrt(len(x))
print('RMSE =', rmse)
# plt.plot(x, y, 'o', label='Original data', markersize=10)
x2 = np.linspace(start = -0.9, stop = 0.5, num = 141)
plt.figure(0)
plt.scatter(x, y, edgecolors='none', s=4, alpha=0.5)
plt.plot(x2, m*x2 + c, 'r', label=('Fitted line, R^2 = %s' % r2))
plt.legend()
plt.show()
```
