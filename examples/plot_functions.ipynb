```python
%matplotlib notebook
import math
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    a = []
    r0 = 5.0
    for item in x:
        a.append(1/(1+math.exp(-(item-r0))))
    return a

def sigmoid2(x):
    a = []
    r02 = 25.0
    for item in x:
        a.append(1/(1+math.exp(-1.0*(item*item - r02))))
    return a

def f_damp(x):
    a = []
    r0 = 4.5
    for item in x:
        a.append(1/(1+math.exp(-20.0*(item/r0 - 1.0))))
    return a
     
x = np.arange(4., 6., 0.01)
sig = sigmoid(x)
sig2 = sigmoid2(x)
damp = f_damp(x)

plt.plot(x,sig)
plt.plot(x,sig2)
plt.plot(x, damp)
plt.show()
```
